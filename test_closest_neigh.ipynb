```python
CALIFA_data = get_data()
data = clean_data_incl(CALIFA_data, 70)
OIII4363_snr = CALIFA_data.loc[:,'fluxOIII4363']/CALIFA_data.loc[:,'e_fluxOIII4363']
mask_OIII4363 = OIII4363_snr > 3
CALIFA_data_clean = CALIFA_data[mask_OIII4363]
```

```python
X = np.log10(CALIFA_data_clean.loc[:,['N2Ha_dd', 'O3Hb_dd']])
X = X.to_numpy()
```

```python
fig, ax = plt.subplots()
ax.scatter(X[:,0],
           X[:,1], c='r', s=20, marker='*',
           alpha=0.5)
```

```python
from scipy.spatial import KDTree, cKDTree
ctree = KDTree(X)
```

```python
inds = ctree.query_ball_point(np.array([[-0.5,0]]), 0.2)
inds = inds[0]
```

```python
inds
```

```python
fig, ax = plt.subplots()
ax.scatter(X[:,0], X[:,1])
ax.scatter(X[inds,0], X[inds,1], c='r')
ax.scatter(-0.5, 0, marker='*', c='k')

```
