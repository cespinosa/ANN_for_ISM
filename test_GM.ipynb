```python
import os
import pandas as pd
import numpy as np
from ANN_manager import read_ANN
from genetic_method import create_parents, create_model
from observational_data import read_Marino_data 
import matplotlib.pyplot as plt
```

```python
ANN = read_ANN('ANN_BOND_ALL',
                os.path.expanduser('~/GoogleDrive/cespinosa/data/ANNs/'))
```

```python
data = create_parents(ANN, 20)
```

```python
data
```

```python
for col in data.columns:
    print(f'{col}\t')
```

```python
obs_data = read_Marino_data()
```

```python
obs_data
```

```python
X = obs_data.loc[:, ['lN2', 'lO3', 'OIIHb', 'OH', 'O3', 'N2', 'O32']].to_numpy()
```

```python
from scipy.spatial import KDTree
tree = KDTree(X)
```

```python
fr = 'high'
age = 1
data_model = data.loc[10, [f'N2_{fr}_{age}', f'O3_{fr}_{age}', f'O2_{fr}_{age}', 
                           'OH_tab', f'rO3_{fr}_{age}', f'rN2_{fr}_{age}']]
```

```python
data_model
```

```python
data_model[f'N2_{fr}_{age}'] = data_model[f'N2_{fr}_{age}'] - np.log10(2.86)
data_model['OH_tab'] = data_model['OH_tab']
data_model['O32'] = np.log10(10**data_model[f'O2_{fr}_{age}'] 
                             + 10**data_model[f'rO3_{fr}_{age}'])
data_model_np = data_model.to_numpy()
```

```python
data_model
```

```python
len(data_model['OH_tab'])
```

```python
distances = []
indices = []
for i in np.arange(100):
    dists, inds = tree.query(np.array([data[i] for data in data_model_np]), 30)
    distances.append(dists)
    indices.append(inds)
```

```python
data_point = np.array([data[99] for data in data_model_np])
fig, axes = plt.subplots(3, 2)
axes[0][0].scatter(X[:,0], X[:,1])
axes[0][0].scatter(X[indices[99],0], X[indices[99],1], c='r')
axes[0][0].scatter(data_point[0], data_point[1], marker='*', c='k')
axes[0][1].scatter(X[:,3], X[:,4])
axes[0][1].scatter(X[indices[99],3], X[indices[99],4], c='r')
axes[0][1].scatter(data_point[3], data_point[4], marker='*', c='k')
axes[1][0].scatter(X[:,3], X[:,5])
axes[1][0].scatter(X[indices[99],3], X[indices[99],5], c='r')
axes[1][0].scatter(data_point[3], data_point[5], marker='*', c='k')
axes[1][1].scatter(X[:,4], X[:,6])
axes[1][1].scatter(X[indices[99],4], X[indices[99],6], c='r')
axes[1][1].scatter(data_point[4], data_point[6], marker='*', c='k')
axes[2][0].scatter(X[:,5], X[:,6])
axes[2][0].scatter(X[indices[99],5], X[indices[99],6], c='r')
axes[2][0].scatter(data_point[5], data_point[6], marker='*', c='k')
```
